```python
import pandas as pd
```

```python
df = pd.read_csv('financials.csv')
df
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df["id"] = df[['Segment', 'Country', 'Product', 'Date', 'Discount Band']].apply(lambda x:"-".join(x.astype(str).str.strip()).lower(), axis=1)
df["id"] = df['id'].str.replace('/','', regex=True)
df.insert(0, 'id', df.pop('id'))
df
```

```python
numerik_df = df[['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts','Sales', 'COGS', 'Profit']]
for x in numerik_df:
    if df[x].dtype == 'object':
        df[x] = df[x].str.replace('$', '').str.replace('-', '0').str.replace(',','').str.strip().str.strip('()')

df
```

```python
df[['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts','Sales', 'COGS', 'Profit']] = df[['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts','Sales', 'COGS', 'Profit']].astype(float)
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed')
df
```
